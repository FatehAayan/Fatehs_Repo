```python
class university:
    campus=5
    def __init__(self,name,address):
        self.name=name
        self.address=address
    def _display(self):
        print(f"{self.name} university is located in {self.address}")
```

```python
class department(university):
    def __init__(self,pname,program,u_name,u_address):
        super().__init__(u_name,u_address)
        self.pname=pname
        self.program=program
    def display(self):
        self._display()
        print(f"Name of Department : {self.pname} | Program : {self.program}")
```

```python
U_Fast=university("FAST", "Islamabad")
U_Fast._display()
```

```python
obj2=department("CS","Ai","FAST","Lahore")
obj2.display()
```
